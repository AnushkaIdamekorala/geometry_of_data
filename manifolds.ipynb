1.We need to check to see if the collection B satisfies the two requirements in the
definition of a basis, items (1) and (2) from Definition. 
As it is easy to see condition (1) satisfies let us forcus how  condition (2) fails.
For that, consider B1 = [a, b] and B2 = [b, c]. The intersection is
B1 ∩ B2 = {b}. In order for condition (2) to be satisfied, we need to find a third basis element B3 = [d, e] such that 
b ∈ [d, e] ⊂ {b} = B1 ∩ B2. This is not possible, as any closed interval
containing b will contain points other than b, hence will not be a subset of B1 ∩ B2.

```python
def randomNormalPoint(n):
    s = np.random.normal(0, 1, n)
    norm = s / np.linalg.norm(s)
    return norm;

def distanceCollection(n):
    list =[]
    distance = []
    for x in range(10000):
        norm = randomNormalPoint(n) 
        distance.append(np.arcsin(abs(norm[0])))
    return distance;

histograms = []
nList = [2,4,10,100]
for n in nList:
    histograms.append(distanceCollection(n))

bins = np.linspace(0, 1.6, 20)
plt.figure(figsize=(15,5)) 
plt.hist(histograms, bins, label=['2','4','10', '100'])
plt.legend(loc='upper right')
plt.ylabel("Count of points")
plt.xlabel("Arc distance to equator")
plt.show()
    
```

```python
## To take the arc distance direct distance is taken between points and sin inverse is taken for the half of that circle.
def randomDistanceGenerator(n):
    distance = []
    for x in range(1000):
        s1 = np.random.normal(0, 1, n)
        norm1 = s1 / np.linalg.norm(s1)
        s2 = np.random.normal(0, 1, n)
        norm2 = s2 / np.linalg.norm(s2)
        dist = np.linalg.norm(norm1 - norm2)
        distance.append(2*np.arcsin(dist/2))
    return distance

histograms = []
nList = [2,4,10,100]
for n in nList:
    histograms.append(randomDistanceGenerator(n))


    
bins = np.linspace(0, 3.5, 30)
plt.figure(figsize=(15,5)) 
plt.hist(histograms, bins, label=['2','4','10', '100'])
plt.ylabel("Count of points")
plt.xlabel("Arc distance between points")
plt.legend(loc='upper right')
plt.show()
```

i According to the observation. When d increases the values of the angular distance(which is same as the distance <br>
through the sphere) between two random points concentrated more to pi/2 when  more the d increases.<br>
ii Rather than equally distributing, when d increases, points are more concentrated towards specific low dimension spheres.
